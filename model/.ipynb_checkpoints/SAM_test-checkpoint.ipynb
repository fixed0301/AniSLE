```python
import json
import cv2
import numpy as np
import matplotlib.pyplot as plt
from typing import List, Tuple, Dict
from segment_anything import SamPredictor, sam_model_registry

def load_image(path: str) -> np.ndarray:
    """이미지 로드 (RGB 형식으로)"""
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img

def load_pose_json(path: str) -> dict:
    """포즈 JSON 파일 로드"""
    with open(path, 'r', encoding='utf-8') as f:
        return json.load(f)

def init_sam(model_type: str = "vit_h", checkpoint: str = None, device="cuda"):
    """SAM 모델 초기화"""
    if checkpoint is None:
        raise ValueError("Please provide a SAM checkpoint path.")
    sam = sam_model_registry[model_type](checkpoint=checkpoint)
    sam.to(device=device)
    predictor = SamPredictor(sam)
    return predictor

def debug_json_structure(pose_dict: dict, label: str):
    """JSON 구조 디버깅"""
    print(f"\n=== {label} JSON Structure ===")
    print(f"Type: {type(pose_dict)}")
    
    if isinstance(pose_dict, dict):
        print(f"Keys: {list(pose_dict.keys())}")
        for key, value in pose_dict.items():
            if isinstance(value, list):
                print(f"  {key}: List with {len(value)} elements")
                if len(value) > 0:
                    print(f"    First element type: {type(value[0])}")
                    if isinstance(value[0], list):
                        print(f"    First element: {value[0][:3] if len(value[0]) > 3 else value[0]}")
                    else:
                        print(f"    First element: {value[0]}")
            else:
                print(f"  {key}: {type(value)}")
    elif isinstance(pose_dict, list):
        print(f"Root list with {len(pose_dict)} elements")
        if len(pose_dict) > 0:
            print(f"First element: {pose_dict[0]} (type: {type(pose_dict[0])})")

def extract_keypoints_from_dict(pose_dict: dict, image_size: Tuple[int,int]) -> List[Tuple[float, float, float]]:
    """JSON에서 키포인트 추출 (수정된 버전)"""
    H, W = image_size
    
    debug_json_structure(pose_dict, "Parsing")

    # Case 1: Root level list (aligned keypoints)
    if isinstance(pose_dict, list):
        print(f"Processing root list with {len(pose_dict)} keypoints")
        kps = []
        for i, pt in enumerate(pose_dict):
            if isinstance(pt, list) and len(pt) >= 2:
                x, y = pt[0], pt[1]
                v = pt[2] if len(pt) > 2 else 1.0
                # 픽셀 좌표를 정규화
                kps.append((x/W, y/H, v))
        print(f"Extracted {len(kps)} keypoints from root list")
        return kps

    # Case 2: Complex structure with bodies (original keypoints)
    if isinstance(pose_dict, dict):
        
        # bodies 키가 있는 경우
        if 'bodies' in pose_dict:
            bodies = pose_dict['bodies']
            print(f"Bodies: {type(bodies)}, length: {len(bodies) if hasattr(bodies, '__len__') else 'N/A'}")
            
            if isinstance(bodies, list) and len(bodies) > 0:
                # bodies는 이미 키포인트 리스트들의 배열
                # 첫 번째 사람의 키포인트 가져오기
                person_keypoints = bodies  # bodies 자체가 키포인트 리스트
                print(f"Found {len(person_keypoints)} keypoints in bodies")
                
                kps = []
                for i, pt in enumerate(person_keypoints):
                    if isinstance(pt, list) and len(pt) >= 2:
                        x, y = pt[0], pt[1]
                        v = pt[2] if len(pt) > 2 else 1.0
                        kps.append((x, y, v))
                        
                print(f"Extracted {len(kps)} keypoints from bodies")
                return kps
        
        # keypoints 키가 있는 경우
        if 'keypoints' in pose_dict:
            keypoints = pose_dict['keypoints']
            print(f"Found keypoints key with {len(keypoints) if hasattr(keypoints, '__len__') else 'N/A'} elements")
            
            if isinstance(keypoints, list):
                # Flat list [x1,y1,v1, x2,y2,v2, ...]
                if len(keypoints) % 3 == 0:
                    kps = [(keypoints[i], keypoints[i+1], keypoints[i+2]) for i in range(0, len(keypoints), 3)]
                    return kps
                # List of [x,y,v] or [x,y]
                elif all(isinstance(pt, list) for pt in keypoints):
                    kps = []
                    for pt in keypoints:
                        if len(pt) >= 2:
                            x, y = pt[0], pt[1]
                            v = pt[2] if len(pt) > 2 else 1.0
                            kps.append((x, y, v))
                    return kps

    print("No keypoints found with known patterns")
    return []

def get_pose_difference_points(orig_kps: List[Tuple[float, float, float]], 
                              aligned_kps: List[Tuple[float, float, float]], 
                              image_size: Tuple[int,int], 
                              thresh: float = 0.05) -> List[Tuple[int, int]]:
    """포즈 차이가 있는 키포인트들의 픽셀 좌표 반환"""
    H, W = image_size
    diff_points = []
    
    print(f"\n=== Pose Difference Analysis ===")
    print(f"Original keypoints: {len(orig_kps)}")
    print(f"Aligned keypoints: {len(aligned_kps)}")

    min_len = min(len(orig_kps), len(aligned_kps))
    for i in range(min_len):
        (x0,y0,v0), (x1,y1,v1) = orig_kps[i], aligned_kps[i]
        
        if v0 < 0.3 or v1 < 0.3:  # visibility threshold
            continue
        
        # 좌표 정규화 확인 및 픽셀 변환
        if x0 <= 1.0 and y0 <= 1.0:  # normalized coordinates
            x0_px, y0_px = x0 * W, y0 * H
        else:
            x0_px, y0_px = x0, y0
            
        if x1 <= 1.0 and y1 <= 1.0:  # normalized coordinates
            x1_px, y1_px = x1 * W, y1 * H
        else:
            x1_px, y1_px = x1, y1
            
        dx, dy = abs(x0_px - x1_px), abs(y0_px - y1_px)
        distance = np.sqrt(dx**2 + dy**2)
        
        print(f"Keypoint {i}: orig({x0_px:.1f},{y0_px:.1f}) -> aligned({x1_px:.1f},{y1_px:.1f}), distance: {distance:.1f}")
        
        if distance > thresh * min(W, H):  # threshold 조정
            cx, cy = int(x1_px), int(y1_px)
            # 이미지 경계 내에서만 추가
            if 0 <= cx < W and 0 <= cy < H:
                diff_points.append((cx, cy))
                print(f"  -> Added as difference point")
    
    print(f"Total difference points: {len(diff_points)}")
    return diff_points

def get_sam_masks_from_points(predictor, image: np.ndarray, points: List[Tuple[int, int]]) -> np.ndarray:
    """SAM을 사용해 포인트들로부터 마스크 생성"""
    if len(points) == 0:
        print("No points provided for SAM")
        return np.zeros(image.shape[:2], dtype=bool)
    
    print(f"\n=== SAM Mask Generation ===")
    print(f"Using {len(points)} points: {points}")
    
    # 이미지를 SAM에 설정
    predictor.set_image(image)
    
    # 포인트들을 numpy 배열로 변환
    input_points = np.array(points)
    input_labels = np.ones(len(points))  # 모든 포인트를 positive로
    
    # SAM으로 마스크 예측
    masks, scores, logits = predictor.predict(
        point_coords=input_points,
        point_labels=input_labels,
        multimask_output=True,
    )
    
    print(f"SAM generated {len(masks)} masks with scores: {scores}")
    
    # 가장 높은 점수의 마스크 선택
    best_mask_idx = np.argmax(scores)
    best_mask = masks[best_mask_idx]
    
    print(f"Selected mask {best_mask_idx} with score {scores[best_mask_idx]:.3f}")
    print(f"Mask coverage: {np.count_nonzero(best_mask)} pixels")
    
    return best_mask

def get_sam_masks_grid(predictor, image: np.ndarray, grid_size: int = 32) -> List[np.ndarray]:
    """그리드 방식으로 SAM 마스크들 생성"""
    print(f"\n=== SAM Grid Generation ===")
    print(f"Grid size: {grid_size}x{grid_size}")
    
    image_torch = image.astype(np.uint8)
    predictor.set_image(image_torch)

    h, w = image.shape[:2]
    masks = []

    ys = np.linspace(0, h - 1, grid_size, dtype=int)
    xs = np.linspace(0, w - 1, grid_size, dtype=int)

    points = [[x, y] for y in ys for x in xs]
    points = np.array(points)

    chunk = 512
    for i in range(0, len(points), chunk):
        pts = points[i:i + chunk]
        input_points = pts
        input_labels = np.ones(len(pts), dtype=int)
        masks_out, scores, logits = predictor.predict(
            point_coords=input_points,
            point_labels=input_labels,
            multimask_output=True,
        )
        for mset in masks_out:
            for m in mset:
                mask = m.astype(bool)
                if mask.sum() < 0.001 * h * w:
                    continue
                masks.append(mask)

    # 중복 제거
    unique = []
    seen = set()
    for m in masks:
        ys, xs = np.where(m)
        if len(xs) == 0:
            continue
        bb = (min(xs), min(ys), max(xs), max(ys))
        if bb in seen:
            continue
        seen.add(bb)
        unique.append(m)
    
    print(f"Generated {len(unique)} unique masks from grid")
    return unique

def select_best_mask_for_pose(masks: List[np.ndarray], diff_points: List[Tuple[int, int]]) -> np.ndarray:
    """포즈 차이 포인트들과 가장 잘 맞는 마스크 선택"""
    if len(masks) == 0:
        print("No masks to select from")
        return np.zeros((100, 100), dtype=bool)
    
    print(f"\n=== Mask Selection ===")
    print(f"Selecting from {len(masks)} masks using {len(diff_points)} difference points")
    
    if len(diff_points) == 0:
        best_mask = max(masks, key=lambda x: x.sum())
        print(f"No difference points, selected largest mask with {np.count_nonzero(best_mask)} pixels")
        return best_mask
    
    h, w = masks[0].shape
    pose_mask = np.zeros((h, w), dtype=bool)
    
    # 포즈 차이 포인트들 주변에 작은 마스크 생성
    for x, y in diff_points:
        if 0 <= y < h and 0 <= x < w:
            y1, y2 = max(0, y-15), min(h, y+15)
            x1, x2 = max(0, x-15), min(w, x+15)
            pose_mask[y1:y2, x1:x2] = True

    best_mask, best_iou = None, 0.0
    for i, m in enumerate(masks):
        inter = np.logical_and(m, pose_mask).sum()
        union = np.logical_or(m, pose_mask).sum()
        if union == 0:
            continue
        iou = inter / union
        coverage = np.count_nonzero(m)
        print(f"  Mask {i}: IoU={iou:.3f}, coverage={coverage} pixels")
        if iou > best_iou:
            best_iou, best_mask = iou, m
    
    if best_mask is None:
        best_mask = max(masks, key=lambda x: x.sum())
        print(f"No good IoU match, selected largest mask")
    else:
        print(f"Selected mask with IoU={best_iou:.3f}")
    
    return best_mask

def visualize_debug_steps(image: np.ndarray, 
                         orig_kps: List[Tuple[float, float, float]], 
                         aligned_kps: List[Tuple[float, float, float]], 
                         diff_points: List[Tuple[int, int]],
                         sam_masks: List[np.ndarray] = None,
                         final_mask: np.ndarray = None):
    """디버그 단계별 시각화"""
    if sam_masks and len(sam_masks) > 0:
        fig, axes = plt.subplots(3, 3, figsize=(18, 15))
    else:
        fig, axes = plt.subplots(2, 2, figsize=(12, 10))
        
    H, W = image.shape[:2]
    
    # 원본 이미지 + 원본 포즈
    axes[0,0].imshow(image)
    for i, (x, y, v) in enumerate(orig_kps):
        if v > 0.3:
            if x <= 1.0 and y <= 1.0:
                x, y = x * W, y * H
            axes[0,0].plot(x, y, 'ro', markersize=8)
            axes[0,0].text(x+3, y+3, str(i), color='red', fontsize=8)
    axes[0,0].set_title(f'Original Pose ({len(orig_kps)} keypoints)')
    axes[0,0].axis('off')
    
    # 정렬된 포즈
    axes[0,1].imshow(image)
    for i, (x, y, v) in enumerate(aligned_kps):
        if v > 0.3:
            if x <= 1.0 and y <= 1.0:
                x, y = x * W, y * H
            axes[0,1].plot(x, y, 'bo', markersize=8)
            axes[0,1].text(x+3, y+3, str(i), color='blue', fontsize=8)
    axes[0,1].set_title(f'Aligned Pose ({len(aligned_kps)} keypoints)')
    axes[0,1].axis('off')
    
    # 차이 포인트들
    axes[1,0].imshow(image)
    for i, (x, y) in enumerate(diff_points):
        axes[1,0].plot(x, y, 'r*', markersize=15)
        axes[1,0].text(x+5, y+5, str(i), color='red', fontsize=10, weight='bold')
    axes[1,0].set_title(f'Difference Points ({len(diff_points)} points)')
    axes[1,0].axis('off')
    
    # 최종 마스크
    if final_mask is not None:
        axes[1,1].imshow(final_mask, cmap='gray')
        axes[1,1].set_title(f'Final SAM Mask ({np.count_nonzero(final_mask)} pixels)')
        axes[1,1].axis('off')
    
        # 추가 시각화 (SAM 마스크들이 있는 경우)
        if sam_masks and len(sam_masks) > 0:
            # 처음 몇 개 SAM 마스크들 보여주기
            for i in range(min(6, len(sam_masks))):
                row = 2 + i // 3
                col = i % 3
                if row < 3:
                    axes[row, col].imshow(sam_masks[i], cmap='gray')
                    axes[row, col].set_title(f'SAM Mask {i} ({np.count_nonzero(sam_masks[i])} px)')
                    axes[row, col].axis('off')
    
    plt.tight_layout()
    plt.show()

def main_sam_mask_generation(image_path: str, 
                           orig_pose_json: str, 
                           aligned_pose_json: str, 
                           sam_checkpoint: str,
                           thresh: float = 0.1,
                           use_pose_points: bool = True):
    """메인 함수: SAM을 사용한 마스크 생성과 시각화"""
    
    # 이미지 로드
    try:
        img_np = load_image(image_path)
        H, W = img_np.shape[:2]
        print(f"Image loaded: {W}x{H}")
    except Exception as e:
        print(f"Error loading image: {e}")
        return
    
    # SAM 초기화
    try:
        sam_predictor = init_sam(model_type='vit_h', checkpoint=sam_checkpoint, device='cuda')
        print("SAM model initialized")
    except Exception as e:
        print(f"Error initializing SAM: {e}")
        return
    
    # 포즈 JSON 로드
    try:
        orig_pose = load_pose_json(orig_pose_json)
        aligned_pose = load_pose_json(aligned_pose_json)
        print("Pose JSONs loaded successfully")
    except Exception as e:
        print(f"Error loading pose JSONs: {e}")
        return
    
    # 키포인트 추출
    try:
        kps_orig = extract_keypoints_from_dict(orig_pose, (H, W))
        kps_aligned = extract_keypoints_from_dict(aligned_pose, (H, W))
        print(f"\nExtracted keypoints:")
        print(f"  Original: {len(kps_orig)} keypoints")
        print(f"  Aligned: {len(kps_aligned)} keypoints")
        
        if len(kps_orig) == 0 or len(kps_aligned) == 0:
            print("ERROR: No keypoints extracted!")
            return
            
    except Exception as e:
        print(f"Error extracting keypoints: {e}")
        return
    
    # 포즈 차이 포인트 추출
    try:
        diff_points = get_pose_difference_points(kps_orig, kps_aligned, (H, W), thresh)
        print(f"Found {len(diff_points)} difference points")
    except Exception as e:
        print(f"Error getting pose difference points: {e}")
        return
    
    # SAM 마스크 생성
    sam_masks_list = None
    try:
        if use_pose_points and len(diff_points) > 0:
            # 포즈 차이 포인트들로부터 직접 마스크 생성
            sam_mask = get_sam_masks_from_points(sam_predictor, img_np, diff_points)
            print("SAM mask generated from pose difference points")
        else:
            # 그리드 방식으로 마스크들 생성 후 최적 선택
            sam_masks_list = get_sam_masks_grid(sam_predictor, img_np, grid_size=16)
            sam_mask = select_best_mask_for_pose(sam_masks_list, diff_points)
            print(f"SAM mask selected from {len(sam_masks_list)} grid masks")
            
        print(f"Final mask coverage: {np.count_nonzero(sam_mask)} pixels")
    except Exception as e:
        print(f"Error generating SAM mask: {e}")
        import traceback
        traceback.print_exc()
        return
    
    # 디버그 시각화
    try:
        visualize_debug_steps(img_np, kps_orig, kps_aligned, diff_points, sam_masks_list, sam_mask)
    except Exception as e:
        print(f"Error in visualization: {e}")
        import traceback
        traceback.print_exc()
    
    # 마스크 저장
    try:
        mask_path = image_path.replace('.', '_sam_mask.')
        if mask_path.endswith('.jpeg'):
            mask_path = mask_path.replace('.jpeg', '.png')
        cv2.imwrite(mask_path, (sam_mask.astype(np.uint8) * 255))
        print(f"SAM mask saved to: {mask_path}")
    except Exception as e:
        print(f"Error saving mask: {e}")

# 사용 예시
if __name__ == "__main__":
    # 파일 경로들을 실제 경로로 변경하세요
    image_path = "../flask_app/static/uploads/groundimg.jpeg"
    orig_pose_json = "keypoints.json"
    aligned_pose_json = "keypoints_aligned.json"
    sam_checkpoint = "checkpoints/sam_vit_h_4b8939.pth"
    
    # SAM 마스크 생성 및 시각화 실행
    main_sam_mask_generation(image_path, orig_pose_json, aligned_pose_json, sam_checkpoint,
                           thresh=0.1, use_pose_points=True)
```
